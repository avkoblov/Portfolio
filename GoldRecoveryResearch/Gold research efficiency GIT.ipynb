### Исследование суммарной концентрации всех веществ на разных стадиях: в сырье (rougher.input.feed), в черновом (rougher.output.concentrate) и финальном концентратах (final.output.concentrate).

```python
#в черновом концентрате после флотации

gr_full['rougher_output_metals'] = (gr_full['rougher.output.concentrate_ag'] 
                            + gr_full['rougher.output.concentrate_au']
                            + gr_full['rougher.output.concentrate_pb']
                                       )
```

У значений в черновом и финальном концентратах наблюдаются выбросы в районе 0, которые могут быть обусловлены ошибками измерений или загрузки данных. Удалим эти выбросы. Оставим только значения в пределах 1.5 стандартных отклонений от 1 и 3 квартилей.

```python
def quart_range(df, column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    dfq = df.loc[(df[column] < q3 + 1.5*iqr) & (df[column] > q1 - 1.5*iqr), column]
    return dfq
#часть датафрейма с аномальными выбросами
list_features = ['rougher_input_metals', 'rougher_output_metals', 'final_output_metals']
for column in list_features:
    gr_full[column] = quart_range(gr_full, column)

# Проверяем результат
gr_full.describe()
```

### Функция для расчета метрики качества SMAPE модели 

```python
def calculate_smape(y, pred) -> float:
    smape = round(np.mean(np.abs(y - pred) / ((np.abs(y) + np.abs(pred))/2))*100, 2)
    return smape
```

Отберем только те признаки в обучающей выборке, которые есть и в тестовой. При этом удалим столбец с датой, он не понадобится при обучении модели

```python
X_train = gr_train[train_cols_in_test].drop('date', axis=1)
```
